```python
#Chat GPT
import numpy as np
import matplotlib.pyplot as plt

# Data
arr = np.array([3310, 3355, 3450, 3480, 3490, 3520, 3540, 3550, 3650, 3730, 3925])

# Stats
mean_arr = np.mean(arr)
std_arr = np.std(arr)
z_scores_arr = (arr - mean_arr) / std_arr
outliers_arr = arr[np.abs(z_scores_arr) > 3]
empirical_arr = arr[(arr >= mean_arr - 3 * std_arr) & (arr <= mean_arr + 3 * std_arr)]

# Skewness and kurtosis
skew_arr = ((mean_arr - np.median(arr)) / std_arr)
kurt_arr = np.sum((arr - mean_arr) ** 4) / (len(arr) * (std_arr ** 4)) - 3

# Quartiles
q1 = np.percentile(arr, 25)
q2 = np.median(arr)
q3 = np.percentile(arr, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Box Plot
plt.figure(figsize=(8, 4))
plt.boxplot(arr, vert=False, patch_artist=True)
plt.title('Box Plot with Quartiles and Outliers')
plt.grid(True)

# Annotating Quartiles
plt.text(q1, 1.05, f'Q1: {q1:.0f}', horizontalalignment='center', color='blue')
plt.text(q2, 1.1, f'Q2 (Median): {q2:.0f}', horizontalalignment='center', color='green')
plt.text(q3, 1.05, f'Q3: {q3:.0f}', horizontalalignment='center', color='purple')

# Annotating Outliers (if any)
for outlier in arr[(arr < lower_bound) | (arr > upper_bound)]:
    plt.plot(outlier, 1, 'ro')  # Marking outlier
    plt.text(outlier, 1.02, f'{outlier}', color='red', horizontalalignment='center')

plt.xlabel("Values")
plt.tight_layout()
plt.show()
```

```python
data = {'Marks' : [78,85,92,88,84,78,95,90,92,85]}
df = pd.DataFrame(data)
```

```python
# mean

mean_df = df['Marks'].mean()
print(mean_df)
```

```python
#median_
median_df = df['Marks'].median()
median_df
```

```python
#mode

mode_df = df['Marks'].mode()
print(mode_df)
```

```python
# range 

range_df = df['Marks'].max() - df['Marks'].min()
range_df
```

```python
# iqr
iqr_df = df['Marks'].quantile(0.75) - df['Marks'].quantile(0.25)
print(iqr_df)
```

```python
# variance

variance_df = df['Marks'].var()
variance_df
```

```python
#std

std_dev_df = df['Marks'].std()
std_dev_df
```

```python
#coefficiant

cv_df = (std_dev_df / df['Marks'].mean()) * 100
print(cv_df)
```

```python
# shape skewness

skew_df = df['Marks'].skew()
kurt_df = df['Marks'].kurtosis()
```

```python
# z-scores
df['Z_Score'] = (df['Marks'] - df['Marks'].mean()) / df['Marks'].std()
print(df['Z_Score'])
```

```python
#detecting outliers

outliers_df = df[np.abs(df['Z_Score']) > 3]
```

```python
# emparical rule

mean_df = df['Marks'].mean()
std_df = df['Marks'].std()
compirical_df = df[(df['Marks'] >= mean_df - 3 * std_df) & (df['Marks'] <= mean_df + 3 * std_df)]
```

```python
# Box Plot 
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Marks'])
plt.title("BoxPlot")
plt.show
```
