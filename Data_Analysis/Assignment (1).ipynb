```python
# 18. Outlier Removal from scratch
data = [4, 7, 10, 14, 36, 16, 18, 20, 67, 22, 2]
data_series = pd.Series(data)


# Using IQR method
Q1 = data_series.quantile(0.25)
Q3 = data_series.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_data = data_series[(data_series >= lower_bound) & (data_series <= upper_bound)]
print("Original Data:", data)
print("Outliers Removed:", list(filtered_data))
```

```python
# 19. Finding Covariance between X and Y
X = [2, 4, 6, 8, 10]
Y = [1, 3, 5, 7, 9]

# Convert to numpy arrays
X = np.array(X)
Y = np.array(Y)

# Calculate means
mean_X = np.mean(X)
mean_Y = np.mean(Y)

# Covariance formula: Σ((X - mean_X) * (Y - mean_Y)) / (n-1)
covariance = np.sum((X - mean_X) * (Y - mean_Y)) / (len(X) - 1)
print("\nCovariance between X and Y:", covariance)
```

```python
#24. Do the following pre- processing techniques on the different 10 data sets.
#(a) Data Cleaning (Handling missing value for rows/columns, Handling Duplicates,
#Outliers detection and removal)

import pandas as pd
import numpy as np

# Example dataset (replace with your dataset)
df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D', np.nan, 'B'],
    'Age': [25, 30, 22, np.nan, 28, 30],
    'Salary': [50000, 54000, 50000, 58000, 62000, 54000],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR', 'HR']
})
print("Original Dataset:\n", df)

# Handling Missing Values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Name'] = df['Name'].fillna(df['Name'].mode()[0])

# Handling Duplicates
df.drop_duplicates(inplace=True)

# Outlier Detection & Removal (IQR method on Salary)
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['Salary'] >= lower_bound) & (df['Salary'] <= upper_bound)]

print("\nAfter Data Cleaning:\n", df)
```

```python
#(b) Handling Categorical data
from sklearn.preprocessing import LabelEncoder

# Assuming df from above
label_encoder = LabelEncoder()
df['Department_Encoded'] = label_encoder.fit_transform(df['Department'])

print("\nCategorical Data Encoded:\n", df[['Department', 'Department_Encoded']])
```

```python
#(c) Scaling of the data
from sklearn.preprocessing import StandardScaler

# Scaling Salary column
scaler = StandardScaler()
df['Salary_Scaled'] = scaler.fit_transform(df[['Salary']])

print("\nData after Scaling:\n", df[['Salary', 'Salary_Scaled']])
```

```python
#(d) Data Normalization
from sklearn.preprocessing import Normalizer

# Normalizing Salary column
normalizer = Normalizer()
df[['Salary_Normalized']] = normalizer.fit_transform(df[['Salary']])

print("\nData after Normalization:\n", df[['Salary', 'Salary_Normalized']])
```

```python
#(e) Identity insights which could be drawn from data and demonstrations of the same
print("\nInsights:")
print("Average Salary:", df['Salary'].mean())
print("Highest Salary:", df['Salary'].max())
print("Number of Employees in HR:", sum(df['Department'] == 'HR'))
```

```python
# Unit- 2 Graphical Visualization 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("\nLoading CSV...")
df = pd.read_csv("assignment.csv")
df
```

```python
months = df['month']
sales = pd.to_numeric(df['sales'], errors='coerce') 
```

```python
# --- 1. Bar Chart ---
plt.figure(figsize=(8,5))
plt.bar(months, sales, color='skyblue')
plt.title("Monthly Sales (Bar Chart)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# --- 2. Pie Chart ---
plt.figure(figsize=(6,6))
plt.pie(sales, labels=months, autopct='%1.1f%%')
plt.title("Sales Distribution by Month (Pie Chart)")
plt.show()
```

```python
# --- 3. Scatter Plot ---
plt.figure(figsize=(8,5))
plt.scatter(months, sales, color='red')
plt.title("Sales Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# --- 4. Line Chart ---
plt.figure(figsize=(8,5))
plt.plot(months, sales, marker='o', color='green')
plt.title("Monthly Sales Trend (Line Chart)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# --- 5. Area Chart ---
month_indices = np.arange(len(months))
plt.figure(figsize=(8,5))
plt.fill_between(month_indices, sales, color="lightgreen", alpha=0.5)
plt.plot(month_indices, sales, marker='o', color="green")
plt.title("Monthly Sales (Area Chart)")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(month_indices, months)
plt.show()
```

```python
# --- 6. Histogram ---
plt.figure(figsize=(8,5))
plt.hist(sales, bins=5, color='orange', edgecolor='black')
plt.title("Sales Frequency Distribution (Histogram)")
plt.xlabel("Sales Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# --- 7. Ogive (Cumulative Frequency) ---
sales_sorted = np.sort(sales)
cumulative = np.cumsum(sales_sorted)
plt.figure(figsize=(8,5))
plt.plot(sales_sorted, cumulative, marker='o', color='purple')
plt.title("Ogive of Sales (Cumulative Frequency)")
plt.xlabel("Sales")
plt.ylabel("Cumulative Frequency")
plt.grid(True)
plt.show()
```

```python
# --- 8. Dot Plot ---
plt.figure(figsize=(8,5))
plt.plot(months, sales, 'o', markersize=10, color='blue')
plt.title("Dot Plot of Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
# Descriptive statistics

#Measures of location (mean,median,mode)

sales = df['sales']

mean_sales = sales.mean()
median_sales = sales.median()
mode_sales = sales.mode()[0]

print("Mean Sales:", mean_sales)
print("Median Sales:", median_sales)
print("Mode Sales:", mode_sales)
```

```python
# Measures of variability,
range_sales = sales.max() - sales.min()
variance_sales = sales.var()
std_dev_sales = sales.std()
iqr_sales = sales.quantile(0.75) - sales.quantile(0.25)

print("Range:", range_sales)
print("Variance:", variance_sales)
print("Standard Deviation:", std_dev_sales)
print("Interquartile Range (IQR):", iqr_sales)
```

```python
# Measures of association between two variables

df['profit'] = [50, 80, 30, 100, 120, 90]

# Covariance
cov = np.cov(df['sales'], df['profit'])[0][1]

# Correlation Coefficient (Pearson)
corr = df['sales'].corr(df['profit'])

print("Covariance between Sales and Profit:", cov)
print("Correlation Coefficient:", corr)
```

```python
# Measures of distribution
from scipy.stats import skew, kurtosis

skewness = skew(sales)
kurt = kurtosis(sales)

print("Skewness:", skewness)
print("Kurtosis:", kurt)
```
